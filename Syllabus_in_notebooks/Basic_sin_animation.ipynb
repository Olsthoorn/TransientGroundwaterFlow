```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import pdb
```

```python
plt.style.use('seaborn-pastel') # makes picture nicer

def newfig(title="title", xlabel="xlabel", ylabel="ylabel", xlim=None, ylim=None):
    fig, ax = plt.subplots()
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    if xlim:
        ax.set_xlim(xlim)
    if ylim:
        ax.set_ylim(ylim)
    ax.grid()
    return fig, ax
```

```python
fig, ax = newfig(xlim=(0, 4), ylim=(-2, 2))

# Ensure that the artists are on the axes before animation. Clear their data or else
# The first lines will remain visible on the canvas during the entire animation
line, = ax.plot([], [], lw=3)

x = np.linspace(0, 4, 1000)

# Notice that line is implicitly global. That's ok in Python. You may prefer making
# the artists explicity global. Further arguments of animate may also be passed
# by the fargs argument of FuncAnimation.

def init(): # sets up the frame background. Make sure you have on it everything you need.
    line.set_data([], []) # Don't want the intial line to be visible all the time.
    return line,  # return tuple of artists, with a single artist we have the comma to return a tuple
def animate(it, x): # updates the frame by resetting the data
    y = np.sin(2 * np.pi * (x - 0.01 * it))
    line.set_data(x, y)
    return line, # also wants a tuple of artists back


# frames is an iterable or int, first arg of animate.
# fargs has the other arguments of animate
# interval: on screen interval in ms between successive frames
# blit, means only change pixels that change between frames.
anim = FuncAnimation(fig, animate, init_func=init, frames=200, interval=20,
                     fargs=[x], blit=True, repeat=True)

plt.close(anim._fig) # never shows the above-generated axes and figure

# HTML has output. So you may capture it and just put the output at the last line of the script
# to show the video. (If HTML is not already on the last line.)
out = HTML(anim.to_html5_video())

out # show the video
```

```python
# Save animation to disk. Note that fps has nothing to do with interval.
# It has but just consider the two separate.
# fps = 1000 / interval
# Sometimes showing the video on screen before saving it alters some details of the
# saved mp4 file. You can prevent that by putting the out = HTML(...) linwe after saving
# the video frames to the mp4 files.
anim.save('Basic_sin_animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
```
