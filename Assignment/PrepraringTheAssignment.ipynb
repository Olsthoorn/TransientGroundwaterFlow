```python
import numpy as np
import scipy
from scipy.special import exp1 as W
import matplotlib.pyplot as plt
```

```python
Q = 1200 # m3/d
kD = 900 # m2/d
S = 0.2
xw = 500 # distance from river
yw = 0.
xr = 0.
yr = 0.
yr =np.linspace( -1000, 1000, 200)
t = 1.0 # d
times = np.logspace(0, 2, 21)
r = np.sqrt((xr- xw)**2 + (yr - yw)**2)

u = r**2 * S / (4 * kD * times)

s = Q/(4 * np.pi * kD) * W(u)

Qr = Q * np.exp(-u)
qr = Qr / (2* np.pi * r)  # specific discharge m2/d through the ring
qx = qr * (xw-xr)/r

qriv =2* qx #totalflux at yr perpendicular to the shore of the river

plt.plot(times, s)
plt.show()

plt.plot(yr, qriv)
dy = yr[1:] - yr[:-1]
dy = np.diff(yr)
qivc = 0.5 *(qriv[:-1] + qriv[1:]) 

Qriv = np.sum(qrivc * dy)
```
