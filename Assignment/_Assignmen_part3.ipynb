# An aquifer storage and recovery system to reduce extraction from river during summer

## Problem statement

A water company extracts water from a small river to treat and distribute it as drinking water for the population of a small town nearby. This is not a problem in winter. However, due to growing demand for drinking water and growing environmental concern, extraction has become more and more problematic during summers when the discharge of this small river is at its lowest. The environmental agency has recently even forbidden to further extract water from the river during the summer months.

In order to solve the problem that this causes for the drinking water supply, the drinking water company has suggested an "Aquifer Storage and Recovery" system (these so-called ASR systems are becoming more and more popular). It wants to take in more river water during winter and inject it through a well (or wells) at some distance from the river into the local water-table aquifer, so that this water can be extracted during the next summer. This way, no water-intake will be necessary during the summer months.

However, the debate that took off between the water company and the environmental agency focuses on whether, or to what extent, this ASR system really makes sense. Can you really store the water in winter and extract it during summer without substantially affecting the already low summer-discharge of the river? Will not much of the stored water flow back to the river through the aquifer during winter? And would the extraction not induce an infiltration from the river into the aquifer during summer, so that there is still a water intake, the only differece now being that it will be invisible?

It's your task as hydrologist of the environmental agency to answer this question and illustrate it quantitatively and also explain it clearly. Your explanation should include why and how you derived your answer. It should show the math and the code.

It is obvious that ASR will work if the distance between the well and the river is large enough. But how large is large enough, and on what does it depend? The water company suggested a distance of 500 m from the river. Should the environmental agency agree?

# Solution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expi
def W(u): return -expi(-u) # Theis well function

def aplot(x, y, linspec='b-', label='label', **kwargs):
    '''Convenience function for plotting, returns axes
    paramters:
    ----------
    x, y:     np.array, line to be plotted
    linspec:  color-line specification like 'r', 'r-', 'bo--' etc
    label:    label forthe line
    **kwargs: other keyword arguments are passed on toe ax.set(...)
    returns:
    --------
    axes
    '''
    ax = plt.figure().add_subplot(111)
    ax.set(**kwargs)
    ax.plot(x, y, linspec, label=label)
    return ax
```

```python
dpm = 30.5 # days per month
kD  = 900 * dpm # m2/month, transmissivity
S   = 0.2 # specific yield
Q   = 1500 * dpm # m3/month, demand

times = np.linspace(0, 72., 721) # [months] 5 years = 60 months, start october 1, = after end of 9th month

# Times [in months] when injection starts, stops and when extraction starts, stops
tws = np.arange(5) * 12 + 9  # injection winter starts [months] )
twe = tws + 6 # injection winterstops [months]
tss = twe + 2 # extraction summer starts [months]
tse = tss + 3 # extraction summer stops [months]

print('Start injection : ',tws, ' [months]')
print('Stop  injection : ',twe, ' [months]')
print('Start extraction: ',tss, ' [months]')
print('Stop  extraction: ',tse, ' [months]')
```

```python
def dd_sum(Q, kD, S, xw, x0, y0, times, tstart):
    """Compute the drawdown due to well at (xw, 0)  + mirror at point (x, y)
    parameters:
    -----------
    tstart : times when Q starts
    times: times, array, [months]
    xw : float, x-coordinate of well, y coordinate = 0
    x0, y0 : float, x and y coordinate of observation point
    returns:
    --------
    drawdown of well + its opposite mirror for all times    
    """
    s_total = np.zeros(times.shape) # initialize
    r2_1 = (+xw-x0)**2 + y0**2 # r**2 of well
    r2_2 = (-xw-x0)**2 + y0**2 # r**2 mirror well
    for tst in tstart:
        t = times - tst
        U1 = r2_1 * S / (4 * np.pi * kD * t[t>0])  # U of well
        U2 = r2_2 * S / (4 * np.pi * kD * t[t>0])  # U of mirror
        s = Q / (4 * np.pi * kD) * (W(U1) - W(U2)) # sum of well + its mirror
        s_total[t>0] += s # superposition
    return s_total
```

```python
xw = 500          # position of well (xw, 0)
x0 = xw; y0 =10.  # position of observation point (x0, y0)

# Superimpose all injections and extractions
s = np.zeros(times.shape) # initialize
s += dd_sum( +0.5 * Q, kD, S, xw, x0, y0, times, tws) # injections
s += dd_sum( -0.5 * Q, kD, S, xw, x0, y0, times, twe) # injection stops
s += dd_sum( -1.0 * Q, kD, S, xw, x0, y0, times, tss) # extractions
s += dd_sum( +1.0 * Q, kD, S, xw, x0, y0, times, tse) # extraction stops

# show it
ax = aplot(times, s, 'r', '', xlabel='times [M]', ylabel='elevation [m]', title='Effect of well + its opposite mirror')
plt.show()
```

The graph shows the head change at a given point over time [months]. The time axis starts at January 1 of the first year when the system is launched with the first 6-months injection period from October 1 of year 1 till April 1 of year 2. The injection if followed by 2 months of rest, after which extraction starts on June 1 of year 2 and lasts until September 1 of year 2. The second full year cycle starts on October 1 of year 2. And so on. Notice that the injection flow is 0.5 Q while the extraction flow is 1.0 Q, but at half the duration of the injection flow.

# Exchange with the river

To see to what extent the injection and extraction affect the river (during summer), we should quantify the flux exchange between the aquifer and the river. We apply the Theis solution, and from it derive the flux. First the flux on a circle at distance $r$ around a single well, then that along the river due to a well and its opposite mirror.

The river is simplified to a straight line, and is a fixed-head boundary. If indeed the river bottom is clogged, and it, therefore, has a resistance against flow, the interaction between river and aquifer is less. Therefore, an analysis in which we assume a direct contact between river and aquifer is adequate as it does not underestimate the impact of the planned system on the river.

To simulate the effect of a well along a (straight) river, we have to use a mirror well at the opposite side of the river. It has to have opposite sign as well to keep the head zero exactly between the well and its mirror.

First compute the flow across a circle of radius $r$ around the well. For this we take the mathematical expression of the Theis well formula and then take its derivative with respect to $r$:

$$  s = \frac {Q_0} {4 \pi kD} \intop_u^\infty \frac {e^{-y}} y dy, \,\,\,\,\, u = \frac {r^2 S} {4 kD t} $$

$$ Q(r, t) = -2 \pi r kD \frac {\partial s} {\partial r} = - 2 \pi r kD \frac {\partial s} {\partial u} \frac {\partial u} {\partial r} $$ 

$$ Q(r, t) = 2 \pi r kD \frac {Q_0} {4 \pi kD} \frac {e^{-u}} u \frac {2 r S} {4 kD t} = Q_0 e^{-u}  $$

So, the total flow $Q(r, t)$ [L3/T] across a ring at distance $r$ around the well is given by the previous expression, where $r$ is hidden in parameter $u$.

Therefore, the specific discharge $q$ [L2/T] is

$$ q(r, t) = \frac {Q_0} {2 \pi r} e^{-u}  $$

For specific discharge at a point along the river with coordinates $(0, y)$ when $(x_w, 0)$ is the coordinate of the well ($x_w$ the distance to the river), the flux perpendicular to the river is

$$ q(y,t) = q(r, t) \frac {x_w} r $$

Because, we have the mirror well, the total flux at this point is twice as much

$$ q(y, t) = 2 q(r,t) \frac {x_w} r $$

$$ q(y, t) = 2 \frac {x_w} r \frac {Q_0} { 2 \pi r} e^{-u} =  \frac {x_w} {\pi r^2} Q_0 e^{-u}$$

The last expression is the specific discharge into the river at coordinate $(0, y)$ when $Q_0$ is injection.

By integrating this expression along $y$, we obtain the total flow into the river.

Although this integration should be over $-\infty<y<\infty$ we integrate over $-a\, x_w < y < a\, x_w$ where a is a constant that we may vary. Intuitiively $a$ should be larger than 5 in order not to miss too much water.

We verify the required value of a by simulating a single well and its mirror running at a constant flow and checking that the computed inflow into the river matches the injection on the long run.


```python
times = np.logspace(0, 6, 61)
xw = 500. # x-coordinate of well
Np  = 500  # number of points along the river

# points along the river between -5xw<y<+5xw
a = 50
ymax = a * xw
y = np.hstack((-np.logspace(0, np.log10(ymax), Np)[::-1],
                np.logspace(0, np.log10(ymax), Np)))


def fqriv(Q, xw, y, kD, S, t):
    """returns flux into the river at points (0, y), due well and its mirror
    and the integrated flow into the river.
    
    parameters:
    -----------
    Q : float, injection [L3/T] (extraction will be negative)
    xw: x-coordinate [L] of well, river is at (x=0, y) 
    y: ndarray, [L] vector of y coordinates
    t: time since injection Q started [T]
    returns:
    --------
    qxriv: ndarray, [L2/T] spec. disch. into river at all y coordinates
    Qxriv: float [L3/T] total discharge into the river from aquifer
    """
    r2 = xw**2 + y**2 # squared distance from well (or mirror well) to river point y)
    U  = r2 * S/(4 * kD * t) # U for well and mirror are the same
    
    # flux from aquifer into river. Factor 2 due to mirror well.
    qxriv = xw/r2/np.pi * Q * np.exp(-U)
    
    # Integrate flux qrivy along river to get total loss toward river
    # Apply Simpson's trapezium rule
    dy = np.diff(y)
    Qxriv = 0.5 * (np.sum(qxriv[:-1] * dy) + np.sum(qxriv[1:] * dy))
    return qxriv, Qxriv


# intialize arrays Qxriv and qxriv to store results for all times
Qxriv = np.zeros(len(times)) # vector [Nt]
qxriv = np.zeros((len(y), len(times))) # 2d array [Ny,Nt]

# the fill them in a loop
for it, t in enumerate(times):
    qxriv[:,it], Qxriv[it] = fqriv(Q, xw, y, kD, S, t) 


# plotting the results
ax1 = aplot(times, Qxriv, 'r', 'Loss to river',
            xlabel='times [months]',
            ylabel='Q_to_riv [m3/month]',
            title='Total loss to river [m3/month] from the aquifer due to well', xscale='log')
ax1.plot(times[[0,-1]], [Q, Q], label='Demand')
ax1.legend(loc='best', fontsize='small')
plt.show()


# Also plot the specific discharge along the river for different times
ax2 = aplot([], [], '', '',
            xlabel='y [m]',
            ylabel='influx [m2/month]',
            title='Specific inflow from river due to well for different times', xlim=(-1000, 1000))
for it, t in enumerate(times):
    ax2.plot(y, qxriv[:,it])
plt.show()


# Verify Q by comparing the injection with the loss to the river
print(Q, Qxriv[-1])
```

The figure shows that it takes a very long time before final equilbrium is reached and 6 months to reach about 50%. The graph, therefore, shows that there will always be a substantial loss of the injected water towards the river using the settings proposed by the water company.

# Quantify the actual situation

Now we have everything in place to quantify the actual situation

```python
def rivExchange(Q, xw, y, kD, S, times, tst):
    """Returns the river exchange for all times given Q starts at times[it]
    
    parameters:
    -----------
    times: ndarray [T]
    tst: starting times of the same injection [T] 
    Returns:
    --------
    qx = array [L2/T] of fluxes into the river (for all river points y and all times)
    Qx = vector [L3/T] of flow into the river for all times
    """
    
    qxriv = np.zeros((len(y), len(times)))
    Qxriv = np.zeros(len(times))
    Qinj  = np.zeros(len(times))
    for ts in tst:
        t = times - ts
        for it, t in enumerate(times-ts):
            if t>0:
                qx, Qx = fqriv(Q, xw, y, kD, S, t)
                qxriv[:,it] += qx
                Qxriv[it]   += Qx
                Qinj[it]    += Q
    return qxriv, Qxriv, Qinj


times = np.linspace(0., 72, 721)
qriv_exch = np.zeros((len(y), len(times)))
Qriv_exch = np.zeros(len(times))
Qt = np.zeros(len(times))  # total net recharge / discharge

# injections starting each year on Oct 1
Qinj = 0.5 * Q
qx, Qx, Qin = rivExchange(Qinj, xw, y, kD, S, times, tws)
qriv_exch += qx
Qriv_exch += Qx
Qt        += Qin

# Stop of injections from year 2 March 31
Qinj = -0.5 * Q
qx, Qx, Qin = rivExchange(Qinj, xw, y, kD, S, times, twe)
qriv_exch += qx
Qriv_exch += Qx
Qt        += Qin


# Start extractions from year 2, June 1
Qinj = -Q
qx, Qx, Qin = rivExchange(Qinj, xw, y, kD, S, times, tss)
qriv_exch += qx
Qriv_exch += Qx
Qt        += Qin

# End of extractions from year 2, Aug 31
Qinj = Q
qx, Qx, Qin = rivExchange(Qinj, xw, y, kD, S, times, tse)
qriv_exch += qx
Qriv_exch += Qx
Qt        += Qin


# plotting the results
ax1 = aplot(times, Qriv_exch, 'r', 'Loss to river',
            xlabel='times [months]',
            ylabel='Q_to_riv [m3/month]',
            title='Total inflow [m3/month] from the aquifer due to well')
ax1.plot(times[[0,-1]], [Q, Q], 'g', label='Demand')
ax1.plot(times[[0,-1]], [0, 0], 'c', label='zero impact')
ax1.plot(times, Qt, 'm', label='injection/extraction')


ax1.legend(fontsize='small')
plt.show()


#ax2 = aplot([], [], '', '',
#            xlabel='y [m]',
#            ylabel='influx [m2/month]',
#            title='Specific inflow from river due to well for different times', xlim=(-1000, 1000))
#for it, t in enumerate(times):
#    ax2.plot(y, qriv_exch[:,it])
#plt.show()
```

# Conclusions

The graph shows the demand (green), the injection during winter and the extraction during summer over 5 years (magenta) and the associated exchange with the river (red). Even though the well is at 500 m from the river and the specificy yield is subtantial with its value of 0.2, about half of the injected water is lost to the river before the next extraction period arrives. Also, during the extraction period, there is a substantial induced infiltration from the river into the aquifer. It must thus be concluded that it is simply not true that the water company extracts during summer the water injected during winter without impacting the river discharge. By then end of the three months that intake from the river is forbidden, the induced infiltration is about half the drinking water demand. Therefore, one cannot say that there is no benefit from this ASR system; it at least reduces the water intake during the summer months substantially, but not totally.

Even if we put the wells at 1000 m from the river, there is still induced infiltration during summer, but it is then reduced to about 20% of the demand by the end of the summer extraction period. To see this, simply change xw to 1000 m and compute again.

Interestingly, when xw=2000 m, there is no induced infiltration during summer, in fact, there is then even a small discharge from the aquifer to the river, which is caused by the delay between injection and its effect on the river, being about 6 months. Therefore, with a good design and the correct distance from the river the system may work and may even be beneficial to the river.
