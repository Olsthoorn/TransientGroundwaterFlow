The analytical solution for the diffusion equation with a sine wave at $x=0$ is:

$$ s(x,t) = A \exp( - a x) \sin(\omega t - a x) $$

where the damping a equals:

$$ a = \sqrt{ \frac {\omega S} {2 kD}} $$

we are going to use the following data
1. kD
1. S
1. T, which is the cycle time
1. A, which is the amplitude

```python
import numpy as np
import matplotlib.pyplot as plt

kD = 900 # [m2/d], transmissivity
S = 0.2 # [-], storage coefficient
T = 0.5 # [d], cycle time
A = 1.25 # [m], amplitude
L = 250 # m
omega = 2 * np.pi / T
a = np.sqrt( omega * S / (2 * kD))

x = np.linspace(0, L, 81)
times = np.linspace(0, 3 * T, 21)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(xlabel='x [m]', ylabel='elevation [m]', title='first sine wave')

ax.plot(x, +A * np.exp( - a * x), 'm', linewidth=3, label='Upper envelope')
ax.plot(x, -A * np.exp( - a * x), 'm', lw=3, label='Lower evenlope')

for t in times:
    s = A * np.exp( - a * x) * np.sin( omega * t - a * x)
    ax.plot (x, s, label='time = {:.2f} d'.format(t))    

ax.legend(fontsize='xx-small')
plt.show()
```
