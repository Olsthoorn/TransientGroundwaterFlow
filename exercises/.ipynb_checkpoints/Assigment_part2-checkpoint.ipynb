Imagine you are monitoring groundwater in a country with only production wells and no observation wells. So your only option is to measure water levels in such production wells. Also imagine that, wherever you go, the producion well is continuously running, mostly for years.

>To measure the groundwater and level for your national database, you have three options:  

>1.  Measure the water level in the well while the well continuous to run (assuming that that is possible.  
1.  Stop the well and measure after a given time, say one hour.  
1.  Suspend a pressure gauge in the well (a small device that measure pressure, and stores the data internally, that can later on be read out and processed in the office.

It's reasonable that the change of head obeys the Theis drawdown equation, which for sufficiently small u (i.e. small r and large t) behaves logarithmically according to

$$ s = \frac Q {4 \pi kD} \ln \left( \frac {2.25 kD t} {r^2 S} \right) $$

The drawdown per log-cycle of time is then

$$ \Delta s_{10} = s(r, 10t) - s(r, t) = \frac Q {4 \pi kD} \ln(10) $$

So that

$$ kD = \frac Q {4 \pi \Delta s_{10}} \ln 10 $$

which can be compute as soon as we have determined the drawdown per log cycle from a graph of drawdown versus time with time on a log-axis.

In the case of our monitoring efforts, the well is shut off for the measuring period, which is equivalent to superimposing an extraction flow opposite to that is currently running at the well. The beauty of such a stop test, is that the extraction is exactly known and does not vary with time. Of course, you need to carefully determine the extraction of the running well before it is shut off and also make sure that it has been running at a virtually constant rate for at leas some time up until the start of the pumping test. Also the test should be sufficiently long. It should capture at least to and if possible three log-cycles of time allow unique straight lines through the rising dots on the semlog plot. If we assume that the first minute after ste stop is necessary to completely stop the water flow, then a stop test of 1000 minutes seems ideal, i.e. about the rest of the day until the next morning. But a couple of hours should be considered a minimum.

Another nice consequence of this test is that the transmissivity that is obtained does not depend on the penetration of the well screen, if it is partial or not. The partial penetration does affect the absolute value of the drawdown, as also does any clogging of the well screen and the borehole wall, but does not affect the difference of heads taken at different times with the same extraction.

Clearly the head measured at the end of the stop test is a much better value to indicate the groundwater situation because
- It's closer to the natural groundwater head (i.e. the heads without pumping)
- It misses local pumping effects.
- It misses extra dradowns due to clogging and partial penetration that are mostly unknown and may have a severe influence on the water level in the well.
- It is independent of the actual drawdown in the well (at lest almost, and less so, the longer the stop test lasted)

# Example / test:

Assume the well has been running for a long time, two years say, at Q = 120 m3/h average and after these two years we come by to do the stop test.
Let's simulate the drawdown over the two years until the pumping test has finished.  Then take the stop-test data and analyze them to get the natural head and the transmissivity.


```python
import numpy as np
from scipy.special import expi
import matplotlib.pyplot as plt
```

```python
r0 = 0.25
kD = 900 # m2/d
Sy = 0.2 # [-]
Qh = -120 # m3/h
Qd = 24* Qh
aYear = 365. # d
clog = 0.001 # extra drawdown per m3/d

t0 = 2 * aYear # start of stop test after start of well

def W(u): return -expi(-u)
def U(r=r0, S=Sy, kD=900, t=t):
    return r**2 * S / (4 * kD * t)

t1   = np.logspace( 0, np.log10(t0), 41)   # time till start stop test
s1 =  Qd * clog + Qd / (4 * np.pi * kD) * W(U( t=t1))     # drawdown of well until stop test

tau  = np.logspace(-6, 0, 41) # time during stop test (since start stop test)
s3 =  -Qd * clog -Qd / (4 * np.pi * kD) * W(U(t=tau)) # rise of well during stop test

t2 = t0+tau
s2 =  Qd * clog + Qd / (4 * np.pi * kD) * W(U(t=t0+tau)) # drawdown well during stop test
```

```python
ax1=plt.figure().add_subplot(111)
ax1.set(xlabel='t [d]', ylabel='head [m]', xscale='linear', ylim=(-8., 0.))
ax1.grid(True)

ax1.plot(t1, s1, 'r', label='original well until stop test')
ax1.plot(t2, s2, 'b', label='original during stop test')
ax1.plot(t2, s2+s3, 'g', label='stop test')
ax1.legend(loc='best', fontsize='small')

plt.show()
```

```python
ax2=plt.figure().add_subplot(111)
ax2.set(xlabel='t [d]', ylabel='head [m]', xscale='linear')
ax2.grid(True)

ax2.plot( tau, s2, 'r', label='due to original well')
ax2.plot( tau, s3, 'b--', label='stop test alone')
ax2.plot( tau, s2 + s3, 'g', label='net effect original+stop test')
ax2.legend(loc='best', fontsize='small')
plt.show()
```

```python
ax2=plt.figure().add_subplot(111)
ax2.set(xlabel='t [d]', ylabel='head [m]', xscale='log')
ax2.grid(True)

ax2.plot( tau, s2, 'r', label='due to original well')
ax2.plot( tau, s3, 'b--', label='stop test alone')
ax2.plot( tau, s2 + s3, 'g', label='net effect original+stop test')
ax2.legend(loc='best', fontsize='small')
plt.show()
```

Hence, we obtain the correct kD and the clogging factor back from this test.

The model also shows that the head at the end of the stop test is still at -0.8 m relative to the groundwater head before the original well started. It is not the natural groundwater level in the area, perhaps, but it is far nearer to it than the head in the well while it is pumping.

Furthermore, it can be discussed that after such a long pumping period of the original well, the natural groundwater level in the area has declined, such that the head at the end of the stop test does make more sense than the original head before the well started pumping.
