## Elastic specific storage as a function of porosity and depth

Here I used same lab data bij Van der Knaap (1959) that were presented by Van der Gun (1980). Clearly, there are other sources that present the result of lab measured compressibility. You'll find them in reservoir-engineering handbooks for example, because the relation is important for oil-recovery in reservoir engineering. In the syllabus it only functions as an example.

@ T.N.Olsthoorn, Dec 2020

```python
import numpy as np
import matplotlib.pyplot as plt

def newfig(title=None, xlabel=None, ylabel=None,
                           xscale=None, yscale=None, xlim=None, ylim=None, figsize=(12, 6), size=10):
    fig, ax = plt.subplots()
    fig.set_size_inches(figsize)
    ax.set_title(title, size=size)
    ax.set_xlabel(xlabel, size=size)
    ax.set_ylabel(ylabel, size=size)
    if xscale: ax.set_xscale(xscale)
    if yscale: ax.set_yscale(yscale)
    if xlim: ax.set_xlim(xlim)
    if ylim: ax.set_ylim(ylim)        
    ax.grid()
    
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(size)

    return ax, fig
```

```python
# Compressibilities 

rho, g, m_per_km = 1000, 9.81, 1000

alpha = 4.4e-10 # N/m2, water
beta = lambda eps, z: eps * (3e-11 + 6.6e-11 * (z / m_p_km) ** -0.7) # [m2/N] with z in m below ground surface 
ss   = lambda eps, z: rho * g * (eps * alpha + beta(eps, z))
```

```python
size= 15
ax,_ = newfig("Specific elastic storage coefficient $S_s$ as function of porosity $\epsilon$ and depth $z$",
           r"$S_s = \rho g (\epsilon \alpha + \beta) \,\, [1/m]$", r"$z$ (=depth) [m]", xscale='log', size=size,
             ylim=(300, 0), xlim=(1e-7, 1e-5))

z = np.linspace(0, 300, 100.)[1:]
for eps in [0.10, 0.15, 0.25, 0.35]:
    ax.plot(ss(eps, z), z, label=r"porosity $\epsilon$ = {}".format(eps))
    
ax.grid(True, which='both', axis='both')
```
